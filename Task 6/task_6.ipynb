```python
#data Preprocessing
complaints=complaints[['Product','Complaint ID']]
complaints['Complaint ID']=complaints['Complaint ID'].astype('int')
complaints
```

```python
#lets assign x and y
import numpy as np
x=complaints.iloc[0:4091494,0:1]
print(x)
y=complaints.iloc[:,1].values
print(y.ndim)
```

```python
#turn the categories into numbers because in the above it's defined that sting cannot convert in to float'

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
Features=["Product"]
One_hot=OneHotEncoder()
transformer=ColumnTransformer([("One_hot",One_hot,Features)],remainder="passthrough")
transformed_x=transformer.fit_transform(x)
transformed_x
```

```python
#here we transformed. ie., formated
pd.DataFrame(transformed_x)
```

```python
print(y)
plt.plot(f"{transformed_x},{y}")
plt.title("Product vs Complaint ID")
```
